```r
library(ggplot2)
library(tidyr)
```

```r
# Leer los datos desde el archivo CSV
data <- read.csv("/HAVD_Exo02.csv", header = TRUE)

```

```r
# Mostrar las primeras filas de los datos
head(data)
data <- data[-1, ]
```

```r
nuevos_nombres <- c("Distrito", "Barrio", "Total", paste0("Edad_", 0:100))
colnames(data) <- nuevos_nombres
head(data)
```

```r
# Selecciona las columnas relevantes para la población por edades
edades <- data[, 4:ncol(data)]
edades
```

```r
## Calcula la suma de cada grupo de edad para todos los barrios
total_edades <- colSums(edades)
total_edades
```

```r
# Crea un data frame con los datos necesarios para el gráfico
df <- data.frame(edad = as.numeric(names(total_edades)), poblacion = total_edades)
```

```r
# Extrae los números de los nombres de las edades y cambia el tipo de datos a character
edades_nombres <- gsub("Edad_", "", names(total_edades))
edades_nombres <- as.character(edades_nombres)
print(edades_nombres)
# Crea un data frame con los datos necesarios para el gráfico
df <- data.frame(edad = as.integer(edades_nombres), poblacion = total_edades)

# Muestra el data frame
print(df)
```

```r
# Create a histogram with ggplot2
plot <- ggplot(df, aes(x = factor(edad), y = poblacion)) +
  geom_bar(stat = "identity", fill = "blue", color = "black") +
  labs(title = "Distribution of the population by age",
       x = "Age",
       y = "Population") +
  theme_minimal() +
  theme(plot.title = element_text(size = 16, face = "bold"))

# Save the plot with adjusted width
ggsave("histogram1.png", plot, width = 55, height = 20, units = "cm")
```

```r
# Assign new names to the columns
nuevos_nombres <- c("Distrito", "Barrio", "Total", paste0("Edad_", 0:100))
colnames(data) <- nuevos_nombres
data
```

```r
# Convert data to long format for ggplot and make Edad numeric
data_long <- tidyr::gather(data, key = "Edad", value = "Poblacion", -Distrito, -Barrio, -Total)
data_long$Edad <- as.numeric(gsub("Edad_", "", data_long$Edad))
data_long
```

```r
# Create a new variable for age groups (grouping by 5)
data_long$Age_Group <- cut(data_long$Edad, breaks = seq(0, 100, by = 5), include.lowest = TRUE)
data_long
```

```r
# Create a histogram with ggplot2
library(ggplot2)

plot <- ggplot(data_long, aes(x = Age_Group, y = Poblacion)) +
  geom_bar(stat = "identity", position = "dodge", color = "black") +
  labs(
    title = "Distribution of the population by age",
    x = "Age Group",
    y = "Population",
    caption = "Source: Population data from the city of Barcelona"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 16, face = "bold"),
    axis.text.x = element_text(angle = 45, hjust = 1)  # Rotate x-axis labels for better visibility
  )

# Save the plot with adjusted dimensions
ggsave("histogram_group.png", plot, width = 50, height = 20, units = "cm")
```
